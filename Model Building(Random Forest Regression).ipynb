**Model Building**

```python
class DecisionTreeRegressor:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        
    def fit(self, X, y):
        self.n_features_ = X.shape[1]
        self.tree_ = self._grow_tree(X, y)
        
    def predict(self, X):
        # Returns numpy array
        return np.array([self._predict(inputs) for inputs in X])
        
    def _grow_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        if depth >= self.max_depth or n_samples < 2:
            # Returs a leafnode with predicted target value
            return LeafNode(self._calculate_leaf_value(y))
        # Selects a subset of features    
        feature_indices = np.random.choice(n_features, np.sqrt(n_features), replace=False)
        # Computes the best feature and best threshold
        best_feature, best_threshold = self._best_split(X, y, feature_indices)
        left_indices = X[:, best_feature] < best_threshold
        right_indices = X[:, best_feature] >= best_threshold
        left_subtree = self._grow_tree(X[left_indices], y[left_indices], depth+1)
        right_subtree = self._grow_tree(X[right_indices], y[right_indices], depth+1)
        # Returns a DecisionNode that represents the current node of the tree, with pointers to its left and right subtrees.
        return DecisionNode(best_feature, best_threshold, left_subtree, right_subtree)
        
    def _best_split(self, X, y, feature_indices):
        best_feature, best_threshold = None, None
        # Calculates the variance reduction
        best_variance_reduction = -np.inf
        for feature in feature_indices:
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                left_indices = X[:, feature] < threshold
                # Checks if there are any samples in left and right subsets
                if sum(left_indices) == 0 or sum(~left_indices) == 0:
                    continue
                variance_reduction = self._calculate_variance_reduction(y, left_indices)
                if variance_reduction > best_variance_reduction:
                    best_feature = feature
                    best_threshold = threshold
                    best_variance_reduction = variance_reduction
        return best_feature, best_threshold

    # It calculates the leaf value and return mean of it
    def _calculate_leaf_value(self, y):
        return np.mean(y)
    
    def _calculate_variance_reduction(self, y, left_indices):
        parent_variance = np.var(y)
        # Variance of left child node
        left_variance = np.var(y[left_indices])
        # Variance of right child node
        right_variance = np.var(y[~left_indices])
        # Total number of samples
        n = len(y)
        n_l, n_r = sum(left_indices), sum(~left_indices)
        variance_reduction = parent_variance - (n_l/n * left_variance + n_r/n * right_variance)
        return variance_reduction
    
    def _predict(self, inputs):
        node = self.tree_
        while isinstance(node, DecisionNode):
            if inputs[node.feature] < node.threshold:
                node = node.left
            else:
                node = node.right
        return node.value
        
        
class DecisionNode:
    def __init__(self, feature, threshold, left, right):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        
class LeafNode:
    def __init__(self, value):
        self.value = value


class RandomForestRegressor:
    def __init__(self, n_estimators=10, max_depth=None, max_samples=None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.max_samples = max_samples
        self.trees = []
        
    def fit(self, X, y):
        for _ in range(self.n_estimators):
            tree = DecisionTreeRegressor(max_depth=self.max_depth)
            indices = np.random.choice(len(X), self.max_samples, replace=True)
            tree.fit(X[indices], y[indices])
            self.trees
    def predict(self, X):
        predictions = np.array([tree.predict(X) for tree in self.trees])
        return np.mean(predictions, axis=0)

    def _calculate_oob_score(self, X, y):
        oob_predictions = []
        for i in range(len(X)):
            indices = [j for j in range(len(X)) if j != i]
            oob_X, oob_y = X[indices], y[indices]
            predictions = np.array([tree.predict(oob_X[i:i+1]) for tree in self.trees])
            oob_predictions.append(np.mean(predictions))
        return np.mean((oob_predictions - y)**2)

    def _bootstrap_indices(self, n_samples):
        indices = np.random.randint(0, n_samples, n_samples)
        return indices
    
    def _bootstrap_sample(self, X, y):
        indices = self._bootstrap_indices(len(X))
        return X[indices], y[indices]

    def fit_with_oob_score(self, X, y):
        self.trees = []
        for i in range(self.n_estimators):
            tree = DecisionTreeRegressor(max_depth=self.max_depth)
            bootstrap_X, bootstrap_y = self._bootstrap_sample(X, y)
            tree.fit(bootstrap_X, bootstrap_y)
            self.trees.append(tree)
            oob_score = self._calculate_oob_score(X, y)
            print(f"Tree {i+1}, oob_score = {oob_score}")
        return self

    def predict(self, X):
        # Predict the output for each tree in the forest
        y_pred = np.zeros(X.shape[0])
        for tree in self.trees:
            y_pred += tree.predict(X)
        return y_pred / len(self.trees)

    def mean_absolute_percentage_error(y_true, y_pred):
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
